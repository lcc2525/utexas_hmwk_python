```python
# Dependencies
import pandas as pd
```

```python
# File to load
file_one = "Pokemon.csv"
```

```python
# Read with Pandas
pke_df = pd.read_csv(file_one)
```

```python
# Extract the following columns: 
column_df = pke_df.loc[:,["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
```

```python
# Create a dataframe of the average stats for each type of pokemon.
Aver_statedf = column_df.groupby(['Type 1']).mean()
```

```python
# Calculate the total power level of each type of pokemon by adding all of the stats together.
# Place the results into a new column.

Aver_statedf["Total"] = Aver_statedf.sum(axis=1)
Aver_statedf
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
Total_rankingsdf = Aver_statedf.sort_values('Total', ascending=False)
Total_rankingsdf.to_csv("pokemon_rankings.csv")
```
